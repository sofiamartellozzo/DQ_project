## Accuracy 
= closeness of a value v to the elements of the corresponding definition domain D

```python
def calculate_accuracy(data_imputated, df):    
    true_data = df.copy()
    imputation_data = data_imputated.copy()

    num_rows = true_data.shape[0]
    num_cols = true_data.shape[1]

    count_correct = np.zeros(num_cols-1)

    tot_number_values = num_rows * (num_cols-1)

    for i in range(num_rows):
        #get the row of the respectivelly dataset
        true_row = true_data.iloc[i]
        imp_row = imputation_data.iloc[i]

        for j in range(num_cols-1):
            #check each column if the values matches
            col = columns[j]
            if true_row[col] == imp_row[col]:
                count_correct[j] += 1

    total_correct = count_correct.sum()

    accuracy_label = {}
    for c in range(num_cols-1):
        label = columns[c]
        accuracy_label[label] = count_correct[c]/num_rows
        accuracy_label[label] = '{0:.1f}%'.format(accuracy_label[label]*100)
        #print('The accuracy for ', label, ' is: ', accuracy_label[label])

    accuracy_dataset = total_correct/tot_number_values
    accuracy_dataset = '{0:.1f}%'.format(accuracy_dataset*100)
    #print('The total accuracy of the dataset is: ', accuracy_dataset)
    accuracy_label['total'] = accuracy_dataset
    return accuracy_label
```

```python
labels_acc = {}
label = list(columns[:-1].copy())
label.append('total')

accuracy = pd.DataFrame(index=label, columns=ds)

for i in range(len(datasets_t1)):
    label_acc = calculate_accuracy(datasets_t1[i], df)
    accuracy[ds[i]] = label_acc
    
print('Here the accuracy of the datasets with the imputation performed with the Mode')
accuracy.transpose()
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

model_pipeline = []
model_pipeline.append(SVC())
model_pipeline.append(DecisionTreeClassifier())
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
def make_classification(classes, X_train, X_test, y_train, y_test):
    
    model_list = ['SVM', 'Decision Tree']
    acc_list = []
    precision_list = []
    recall_list = []
    f1_score_list = []
    cm_list = []
    report_list = []

    num_classes = len(classes)

    for model in model_pipeline:
        model.fit(X_train, y_train)

        y_pred = model.predict(X_test)
        acc_list.append(accuracy_score(y_test, y_pred).round(3))
        precision_list.append(precision_score(y_test, y_pred, average='macro').round(3))
        recall_list.append(recall_score(y_test, y_pred, average='macro').round(3))
        f1_score_list.append(f1_score(y_test, y_pred, average='macro').round(3))
        cm_list.append(confusion_matrix(y_test, y_pred))
        report_list.append(classification_report(y_test, y_pred, target_names= classes, output_dict=True))
        
    return acc_list, precision_list, recall_list, f1_score_list, report_list, cm_list

```

```python
acc_list_t1 = {}
precision_list_t1 = {}
recall_list_t1 = {}
f1_score_list_t1 = {}
cm_list_t1 = {}
report_list_t1 = {}

for i in range(len(datasets_t1)):
    acc_list_t1[ds[i]], precision_list_t1[ds[i]], recall_list_t1[ds[i]], f1_score_list_t1[ds[i]], report_list_t1[ds[i]], cm_list_t1[ds[i]] = make_classification(classes, X_train_t1[ds[i]], X_test_t1[ds[i]], y_train_t1[ds[i]], y_test_t1[ds[i]])
    
    
```
